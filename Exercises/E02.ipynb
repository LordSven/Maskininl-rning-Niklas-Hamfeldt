```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 0)

```python
np.random.seed(42)
rng = np.random.rand(1000, 2)
n_rng = np.random.normal(0, 1, 1000)
int_rng = np.column_stack((np.ones(1000), rng))
y = 3*rng[:,0] + 5*rng[:,1] + 3 + n_rng
```

# 1a)

```python
theta = np.zeros(3)
thetas = []

for epoch in range(1, 500):
    theta = theta - (0.1 * (2 / int_rng.shape[0]) * int_rng.T @ (int_rng @ theta - y))
    if epoch % 5 == 1:
        thetas.append(theta.copy())

thetas = np.array(thetas)

epochs = range(0, 500, 5)
plt.plot(epochs, thetas[:, 0], label="theta0")
plt.plot(epochs, thetas[:, 1], label="theta1")
plt.plot(epochs, thetas[:, 2], label="theta2")
plt.legend()
```

# 1b)

```python
theta2 = np.zeros(3)
thetas2 = []

for epoch in range(1, 5000):
    theta2 = theta2 - (0.01 * (2 / int_rng.shape[0]) * int_rng.T @ (int_rng @ theta2 - y))
    if epoch % 20 == 1:
        thetas2.append(theta2.copy())

thetas2 = np.array(thetas2)

epochs2 = range(0, 5000, 20)
plt.plot(epochs2, thetas2[:, 0], label="theta0")
plt.plot(epochs2, thetas2[:, 1], label="theta1")
plt.plot(epochs2, thetas2[:, 2], label="theta2")
plt.legend()
```

```python
SGD_theta = np.zeros(3)
SGD_thetas = []

for epoch in range(1, 500):
    indices = np.random.permutation(1000)
    for i in indices:
        Xi = int_rng[i:i+1]
        yi = y[i:i+1]
        SGD_theta = SGD_theta - (.1 * (2/1) * Xi.T @ (Xi @ SGD_theta - yi))
    if epoch % 5 == 1:
        SGD_thetas.append(SGD_theta.copy())

SGD_theta = np.array(SGD_thetas)

SGD_epochs = range(0, 500, 5)
plt.plot(SGD_epochs, SGD_theta[:, 0], label="theta0")
plt.plot(SGD_epochs, SGD_theta[:, 1], label="theta1")
plt.plot(SGD_epochs, SGD_theta[:, 2], label="theta2")
plt.legend()
```
